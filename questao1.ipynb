**"Para amostras aléatorias simples (X1,X2,X3,..., Xn), retiradas de uma população com média μ e variancia $σ^2$ finita, a distribuição amostral média $\overline{X} = (X1,X2,X3,..., Xn)/n$ aproxima-se de uma distribuição normal com média  μ e variancia $(σ^2)/n$, se n grande"**

Sabendo deste enunciado e observando o gráfico do item b, podemos afirmar que o teorema se faz verdadeiro, uma vez que temos 1000 amostras para o estudo e o valor central do histograma, que se aproxima da média por ter maiores quantidade de dados, também se aproxima do valor central da distribuição normal, que tem sua média no centro e o desvio padrão deslocando a esquerda e direita do centro.
