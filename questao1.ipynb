```python
def gaussiana(x,a,b,c):
    return a*np.exp(-(x-b)**2/(2*c**2))
```

```python
def normal_exponencial(media, n):
    desv_padrão = 1/np.sqrt(n)

    x = np.linspace(media-4*desv_padrão, media+4*desv_padrão, 1000)

    #Funcao densidade de probabilidade

    fdp = (1/(desv_padrão*np.sqrt(2*np.pi)))*np.exp(0.5*((x-media)/desv_padrão)**2)

    return x, fdp
```

```python
x, fdp = normal_exponencial(medias, 1000)
```
