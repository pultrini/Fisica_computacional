```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# F^-1 

def inversa(n, alfa, x_min):
    u = np.random.uniform(0,1,n)
    x = x_min*(u**(1/(-alfa+1)))
    return x
```

```python
def f(x, alfa, x_min):
    y_values = []
    for x_ in x:
        if x_ >= x_min:    
            y = ((alfa-1)*x_min**(alfa-1))*x_**(-alfa)
            y_values.append(y)
        else:
            y_values.append(0)
    return y_values
```

```python

alfa = 2.5
x_min = 1
x_max = 10
n = 10000
x_values = np.linspace(0.0, x_max, 100)


dist_values = inversa(n,alfa,x_min)

y = f(x_values,alfa,x_min)
```

```python
plt.figure(figsize=(8,6))
plt.histist(dist_values, bins=10000, edgecolor = 'black', density=True)
plt.plot(x_values, y, label = 'Curva teorica de f(x)', color = 'red', alpha=1)
plt.xlim(0, x_max)
plt.title('DistribuiÃ§Ã£o Lei de PotÃªncia')
plt.xlabel('Valores de x')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```
