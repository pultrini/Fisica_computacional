```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# F^-1 

def F(u, alfa, x_min):
    x = x_min*u**(-1/(alfa-1))
    return x
```

```python
def f(x, alfa, x_min):
    p1 = x_min**(alfa-1)
    p2 = x**(-alfa+1)
    return p1*p2
```

```python

alfa = 2.5
x_min = 1
x_max = 10
n = 10000
u = np.random.rand(n)
x_values = np.linspace(0, x_max, 1000)


x = F(u,alfa,x_min)
y = f(x, alfa, x_min)
```

```python
#histograma obtido

plt.hist(x, bins=50, range=(0, x_max), edgecolor = 'black')
plt.title('Distribui√ß√£o Lei de Pot√™ncia')
plt.xlabel('Valores de x')
plt.ylabel('Densidade')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(x, bins=50, range=(0, x_max), edgecolor = 'black')
plt.scatter(x, y*2500, label = 'Curva teorica de f(x)', color = 'red', alpha=0.3)
plt.xlim(0, x_max)
plt.text(2.5,2300,"Foi multiplicado f(x) por 2500 para melhor visualiza√ß√£o", style = 'italic')
plt.title('Distribui√ß√£o Lei de Pot√™ncia')
plt.xlabel('Valores de x')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```
